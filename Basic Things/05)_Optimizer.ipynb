```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
model = nn.Linear(1,1)

x = torch.tensor([2.0])
y_true = torch.tensor([5.0])

loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

for i in range(1000) :
    y_pred = model(x)
    loss = loss_fn(y_pred , y_true)
    loss.backward()

    optimizer.step()
    optimizer.zero_grad()

    if i % 10 == 0:
        print(f"Epoch {i}, Loss: {loss.item()}")
```

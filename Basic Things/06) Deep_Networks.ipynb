```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
model = nn.Sequential(
    nn.Linear(1,20),
    nn.ReLU(),

    nn.Linear(20,10),
    nn.ReLU(),

    nn.Linear(10,1),
    nn.Sigmoid()
)

x = torch.tensor([2.0])
y = torch.tensor([1.0])
```

```python
loss_fn = nn.CrossEntropyLoss()
opt = optim.SGD(model.parameters(), lr=0.01)

for i in range(1000):
    output = model(x)
    loss = loss_fn(output  ,y)
    loss.backward()

    opt.step()
    opt.zero_grad()
    if i % 100 == 0:
        print(loss.item())
```

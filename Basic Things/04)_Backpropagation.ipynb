```python
import torch
import torch.nn as nn
```

```python
model = nn.Linear(1,1)

x = torch.tensor([2.0])
y_true = torch.tensor([5.0])
y_pred = model(x)

loss_fn = nn.MSELoss()
loss = loss_fn(y_pred, y_true)

loss.backward()
```

```python
import torch
import torch.nn as nn
```

```python
x = torch.randn(1, 1, 4, 4)

pool =nn.MaxPool2d(2)

print(pool(x))
```

## `pool = nn.MaxPool2d(2)`

- Pool size = 2√ó2
- Takes maximum value in each region
- No parameters to learn

Pooling is deterministic

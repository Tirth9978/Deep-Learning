# Multiple Input Channels
So far, you used 1 channel.
Real images have 3 channels: R, G, B üì∏<br>
for the Input Chanal it will be 3 instaded of 1

````python
import torch
import torch.nn as nn
````

````python
x = torch.randn(1, 3, 5, 5)

conv = nn.Conv2d(3, 2, kernel_size=3, padding=1)

y = conv(x)

print(y.shape)
````

## Note :
üß† Each filter is actually 3√ó3√ó3, not 3√ó3 ‚ùó

- üî• Critical Insight (Most Miss This)
  - You are not applying 3 separate Filters
  - You are applying ONE filter with depth 3
  - That filter looks at all channels together

One output feature map = sum over all channels

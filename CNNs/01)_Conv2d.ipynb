# CNNs

````python
import torch
import torch.nn as nn
````

````python
x = torch.randn(1, 1, 5, 5)

conv = nn.Conv2d(1,2,kernel_size = 3)

y = conv(x)

print(x.shape)
print(y.shape)
````

```python
import torch
import torch.nn as nn
```

```python
class Attn(nn.Module):
    def __init__(self , h):
        super().__init__()
        self.w = nn.Linear(h, 1)

    def forward(self , x) :
        a = self.w(x)
        a = torch.softmax(a, 1)
        out = (a * x).sum(1)       # focus
        return out
```

```python
class RNN_Attn(nn.Module) :
    def __init__(self , v , e , h):
        super().__init__()
        self.emb = nn.Embedding(v , e)
        self.RNN = nn.RNN(e , h , batch_first = True)
        self.attn = Attn(h)
        self.fc = nn.Linear( h , v)

    def forward(self , x):
        x =self.emb(x)
        o, _ = self.rnn(x)
        c = self.attn(o)
        y = self.fc(c)
        return y
```

```python
class Model(nn.Module):
    def __init__(self ):
        super().__init__()
        self.emb = nn.Embedding(v , 10)
        self.rnn =  nn.RNN(10 , 20 , batch_first=True)
        self.attn = Attn(20)
        self.fc = nn.Linear(20 , v)

    def forward(self , x):
        x = self.emb(x)
        o, _ =self.rnn(x)
        c = self.attn(o)
        y=self.fc(c)
        return y
```

```python
txt = "I am Tirth , I am from India"

ch = list(set(txt))
c2i = {c:i for i,c in enumerate(ch)}
i2c = {i:c for c,i in c2i.items()}
v = len(ch)

inp = "I am Tirth , I am from "
tgt = "India"

x = torch.tensor([[c2i[c] for c in inp]])
y = torch.tensor([[c2i[c] for c in tgt]])
```

```python
m = Model()
loss_fn = nn.CrossEntropyLoss()
opt = torch.optim.Adam(m.parameters() , lr=0.01)

for i in range(1000):

    opt.zero_grad()
    output = m(x)

    loss= loss_fn(output , y[:,0])
    loss.backward()

    opt.step()
    if i % 100 == 0:
        print("loss:", loss.item())
```

```python
with torch.no_grad():
    output = m(x)
    p = torch.argmax(output)
    print("Predicted letter:", i2c[p.item()])
```

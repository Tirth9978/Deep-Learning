```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
```

## TRAINING LOOP

```python
def train_simple_model() :
    """Complete training pipeline"""

    # Creating the Data Set
    np.random.seed(42)
    X = np.random.randn(100, 2)
    y = (X[:, 0] + X[:, 1] > 0).astype(float)

    # Converting the Data to PyTorch
    X_tensor = torch.FloatTensor(X)
    y_tensor = torch.FloatTensor(y).unsqueeze(1)

    # ğŸ—ï¸ Create simple model
    model = nn.Sequential(
        nn.Linear(2,4),
        nn.ReLU(),
        nn.Linear(4, 1),
        nn.Sigmoid()
    )

    # ğŸ¯ Define loss and optimizer
    criterion = nn.BCELoss()  # Binary Cross-Entropy
    optimizer = optim.Adam(model.parameters(), lr=0.01)  # âš¡ Adam optimizer

    # ğŸ“ˆ Training tracking
    losses = []

    # Traing Loop
    for epoch in range(100):
        # Forward pass
        predictions = model(X_tensor)
        loss = criterion(predictions, y_tensor)

        # Backward pass and optimization
        optimizer.zero_grad()  # ğŸ§¹ Clear old gradients
        loss.backward()        # ğŸ“‰ Compute gradients (backpropagation)
        optimizer.step()       # ğŸ“ Update weights

        # Track progress
        losses.append(loss.item())

        if epoch % 20 == 0:
            print(f"Epoch {epoch:3d} | Loss: {loss.item():.4f}")

        # ğŸ“Š Plot training loss
    plt.figure(figsize=(10, 4))
    plt.plot(losses)
    plt.title("ğŸ“‰ Training Loss Over Time")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.grid(True)
    plt.show()

    return model, losses


# ğŸƒ Run training
trained_model, loss_history = train_simple_model()
print(f"\nâœ… Training complete! Final loss: {loss_history[-1]:.4f}")
```

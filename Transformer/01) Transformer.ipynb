## What is a Transformer block?
One block contains:
1. 1ï¸âƒ£ Self-Attention
2. 2ï¸âƒ£ Feed-Forward Network

````python
import torch
import torch.nn as nn
````

````python
encode_layer = nn.TransformerEncoderLayer(
    d_model=128,
    nhead=4
)

transformer_encoder = nn.TransformerEncoder(
    encode_layer,
    num_layers=2
)

# Step 3: Dummy input (sequence_length, batch_size, embedding_dim)
x = torch.randn(10, 32, 128)

# Step 4: Forward pass
output = transformer_encoder(x)

print(output.shape)
````

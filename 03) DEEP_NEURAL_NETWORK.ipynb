```python
import torch
import torch.nn as nn
```

## DEEP NEURAL NETWORK

```python
class SimpleDNN(nn.Module) :
    """Deep Neural Network with multiple layers"""

    def __init__(self , input_Size, hidden_sizes , output_size):
        super().__init__()
        layers = []
        prev_size = input_Size

        # Add hidden layers

        for hidden_size in hidden_sizes :
            layers.append(nn.Linear(prev_size , hidden_size))
            layers.append(nn.ReLU())
            layers.append(nn.Dropout(0.3))
            prev_size = hidden_size # Corrected: update prev_size for the next layer

        # Add output layer
        layers.append(nn.Linear(prev_size , output_size))
        layers.append(nn.Sigmoid())

        # # Create sequential model
        self.model = nn.Sequential(*layers)

    def forward(self , X):
        return self.model(X) # Corrected: added return statement

print("Making the Model")
DNN = SimpleDNN(
    input_Size=4 ,
    hidden_sizes=[3,4,2], # 3 hiddel layers
    output_size=1 # Binary classification
)
print("Network Architecture:")
print(DNN)

input = torch.randn(3,4) # Corrected: input tensor shape to match input_Size
output = DNN(input)
output
```

# Import libraries

```python
import torch
import torch.nn as nn
```

```python
class CharRNN(nn.Module) :
    def __init__(self) :
        super().__init__()
        self.embedding = nn.Embedding(vocab_size,embedding_dim=10)
        self.rnn = nn.RNN(
            input_size=10,
            hidden_size=200,
            batch_first=True
        )
        self.fc = nn.Linear(200,vocab_size)

    def forward(self , x) :
        x = self.embedding(x)
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])  # last character output
        return out
```

```python
text = "I am Tirth , I am from India"

chars = list(set(text))
char_to_idx = {ch: i for i, ch in enumerate(chars)}
idx_to_char = {i: ch for ch, i in char_to_idx.items()}
vocab_size = len(chars)

input_text = "I am Tirth , I am from "
target_text = "India"

X = torch.tensor([[char_to_idx[ch] for ch in input_text]])
y = torch.tensor([char_to_idx[ch] for ch in target_text])

model =  CharRNN()
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for i in range(1000) :
    optimizer.zero_grad()
    output = model(X)

    loss = loss_fn(output , y[0].unsqueeze(0))
    loss.backward()
    optimizer.step()

    if i % 50 == 0:
        print(f"Epoch {i}, Loss: {loss.item():.4f}")

with torch.no_grad():
    output = model(X)
    predicted_idx = torch.argmax(output)
    predicted_char = idx_to_char[predicted_idx.item()]
    print("Predicted first letter:", predicted_char)
```

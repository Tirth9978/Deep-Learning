```python
import torch
import torch.nn as nn
```

## DENSE LAYER (Multi-Output Perceptron)

```python
class DenseLayer(nn.Module) :
    """Dense layer: All inputs connected to all outputs"""
    def __init__(self,input_size, output_size) :
        super().__init__()

        self.linear = nn.Linear(input_size , output_size)
        self.activation = nn.ReLU() # âš¡ Can change to any activation

    def forward(self , X) :
        return self.activation(self.linear(X))

dense = DenseLayer(input_size=4, output_size=2)
test_input = torch.randn(3, 4)
output = dense(test_input)
output
```

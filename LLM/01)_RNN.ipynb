```python
import torch
import torch.nn as nn
```

```python
rnn = nn.RNN(
    input_size = 1 ,
    hidden_size=4,
    batch_first=True
)
```

```python
x = torch.tensor([
    [[1.0],
     [2.0],
     [3.0]]
])
```

```python
output, hidden = rnn(x)
```

```python
print("Output : " , output)
print("Hidden : " , hidden)
```

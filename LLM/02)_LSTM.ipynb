```python
import torch
import torch.nn as nn
```

## Create an LSTM layer

```python
lstm = nn.LSTM(
    input_size = 1,
    hidden_size=4 ,
    batch_first=True
)
```

```python
x = torch.tensor([
    [[1.0],
     [2.0],
     [3.0],
     [4.0]]
])
output, (h_n, c_n) = lstm(x)
```

```python
print("Output:", output)
print("Hidden state (h_n):", h_n)
print("Cell state (c_n):", c_n)
```

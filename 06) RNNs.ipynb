```python
import torch
```

```python
class SimpleRNN(torch.nn.Module):
    def __init__(self) :
        super().__init__()

        self.rnn = torch.nn.RNN(
            input_size=1,
            hidden_size=5,
            batch_first=True
        )

        self.fc = torch.nn.Linear(5, 1)

    def forward(self, x):
        x, _ = self.rnn(x)
        x = self.fc(x)
        return x

# Shape: (batch, time_steps, input_size)
x = torch.tensor([[[1.0], [2.0], [3.0], [4.0]]])
model = SimpleRNN()
output = model(x)
print(output)
```
